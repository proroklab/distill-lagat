```python
%load_ext autoreload
%autoreload 2

from pathlib import Path

import numpy as np
import torch

np.set_printoptions(precision=3, suppress=True)
torch.set_printoptions(precision=2, linewidth=200)
```

```python
# load model

from lagat.model import GNNPlanner

model_fpath = Path("../assets/pretrained/loss_best.pt")
model = GNNPlanner.reconstruct(model_fpath)
model.eval()
```

```python
# setup grid config

from pogema import GridConfig

from lagat.utils import set_global_seeds

seed = 42

set_global_seeds(seed)
grid_config = GridConfig(
    num_agents=20,
    obs_radius=model.obs_radius,
    seed=seed,
    observation_type="MAPF",
    on_target="nothing",
    density=0.2,
    width=20,
    height=20,
)
```

```python
# run model

from lagat.run_planners import run_neural_policy

%time env, all_actions, all_observations, all_terminated = run_neural_policy(model, grid_config, animation=True)
```

```python
# visualize

env.save_animation("tmp.svg")
```
